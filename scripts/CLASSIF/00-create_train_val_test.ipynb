### 1.2 Add tags

```python
# Path to a csv containing additionnal metadata
document_path = ROOT + '/data/METADATA/details_archives_94_ETS.csv'

# Check if the document exists
if os.path.exists(document_path):
    print(f"{document_path} exists.")
else:
    print(f"{document_path} does not exist.")
```

```python
metadata = pd.read_csv(document_path)
```

```python
#Filter the columns
metadata_filtered = metadata[["Cote","Tag_groupe_doc","P_debut_sampling","P_fin_sampling","Num_couleur_gris"]]
display(metadata_filtered)
```

```python
# Add additional metadata to the elements df
group = []
color = []
for i in range(len(elements_df)):
  for j in range(len(metadata_filtered)):
    if (elements_df["dossier_cote"].iloc[i] == metadata_filtered["Cote"].iloc[j] and float(elements_df["image_index"].iloc[i]) >= metadata_filtered["P_debut_sampling"].iloc[j] and float(elements_df["image_index"].iloc[i]) <= metadata_filtered["P_fin_sampling"].iloc[j]):
      group.append(metadata_filtered["Tag_groupe_doc"].iloc[j])
      color.append(metadata_filtered["Num_couleur_gris"].iloc[j])

df = elements_df
df["tag"] = group
df["color"] = color
display(df)
```

```python
#Save the dataframe as a csv to ease the following treatements (don't call Arkindex API again)
df.to_csv(path_or_buf=ROOT+'/data/METADATA/pages_annotees.csv',index=False,encoding='utf8',sep='\t')
```

## 2. Elements sampling

```python
# Delete pages without class
print(f"Number of retrieved pages {len(df)}")
df = df[df["classe"] != ""]
print(f"Number of pages with at least one classe {len(df)}")
```

```python
#Define the straficication groups (=create a new column with values equal to the concatenation of the columns to use in the sampling)
df["stratification_id"] = df["classe"]
```

```python
from sampling import stratified_groups

train, val, test = stratified_groups(df, "commune")
```

```python
for sub in [train, val, test]:
  print(sub["color"].value_counts())
```

## 3. Create a gold standard (to used to create YOLOv11 dataset)

```python
# Add a new column to each subset (dataframes) to indicate the subset name
train_json = train.copy()
train_json["subset"] = "train"
val_json = val.copy()
val_json["subset"] = "val"
test_json = test.copy()
test_json["subset"] = "test"
```

```python
#Create a new dataframe (concatenation of the 
gt = pd.concat([train_json,val_json,test_json])
gt = gt.drop('stratification_id', axis=1)
gt = gt.drop('dossier_cote', axis=1)
gt = gt.reset_index()
gt = gt.drop('index', axis=1)
display(gt)
```

```python
result = gt.to_json(ROOT + '/data/CLASSIF/gt_v0.json',orient="records")
```
