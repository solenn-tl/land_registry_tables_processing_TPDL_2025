## 1. Select entries for inference

```python
import pandas as pd
#Import entities
ENTITIES_CSV = ROOT + "/data/NER/entities.csv"
gold_reference = pd.read_csv(ENTITIES_CSV)
display(gold_reference)
```

```python
import json
import glob

json_files = glob.glob(ROOT + "/scripts/NER/*.json")

# Initialize an empty dictionary to store the merged result
ids_to_remove = []

# Loop through the JSON files
for json_file in json_files:
    with open(json_file, 'r') as f:
            data = json.load(f)
            # Merge the current data into the merged_data dictionary
            # You can customize how merging is handled here.
            for elem in data:
                ids_to_remove.append(elem["element_uuid"])
```

```python
filtered_df = gold_reference[~gold_reference["element_uuid"].isin(ids_to_remove)]
```

```python
filtered_df
```

```python
filtered_df = filtered_df[filtered_df['taxpayer'] != '§']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'id']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'idem']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'Id']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'Idem']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'Le même']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'Les mêmes']
filtered_df = filtered_df[filtered_df['taxpayer'] != 'Ø']
filtered_df = filtered_df[filtered_df['taxpayer'] != '"']
filtered_df = filtered_df[filtered_df['taxpayer'] != '\'']
filtered_df = filtered_df.dropna(subset=["taxpayer"])
```

## Inference

```python
### Model to run
MODEL_PATH = OUT_BASE / 'tmp'

FINETUNED_MODEL = MODEL_PATH / "ner-joint-labelling-io-417/checkpoint-600"
FINETUNED_MODEL
```

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForTokenClassification, TokenClassificationPipeline

LIMIT = 100

tokenizer = AutoTokenizer.from_pretrained(FINETUNED_MODEL) #Param : tokenizer du modele souhaité
model = AutoModelForTokenClassification.from_pretrained(FINETUNED_MODEL) #Modèle choisi

#Classification des entités
nlp = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="simple", use_fast=True, ignore_labels=["O+O"])
```

```python
import time
import datetime

results = []
start = time.time()
for ix, l in filtered_df.iterrows():
    res = nlp(l["taxpayer"])
    results.append([l["element_uuid"],l["taxpayer"],res])
end = time.time()
run = end-start
average_runtime = run/len(results)
print(f"Running time for {len(results)} entries is {round(run,3)} seconds (average per entry : {round(average_runtime*1000,3)} milliseconds)")
```

Running time for 1976 entries is 12.375 seconds (average per entry : 6.263 milliseconds)

```python
import spacy
from spacy import displacy
from spacy.tokens import Span

colors = {
    "name": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
    "firstnames": "#ffcc00",
    "address": "#198cdb",
    "activity": "#90e0ef",
    "title": "#ff595e",
    "familystatus": "#31cc27",
    "birthname":"#FFC300"
}

options = {"colors": colors}

for INDEX in range(len(results)):

    # Prepare data for spaCy's displacy
    tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(results[INDEX][0])))
    entities = []
    
    for entity in results[INDEX][2]:
        entities.append({
            "start": entity["start"],
            "end": entity["end"],
            "label": entity["entity_group"].replace('+O','')
        })
    
    # Format as required by displacy
    displacy_data = {
        "text": results[INDEX][1],
        "ents": entities,
        "title": None
    }
    
    # Render visualization
    displacy.render(displacy_data, style="ent", manual=True, options=options)
```

## Convert to JSON

```python
full_json = []
for elem in results:
    entry = {"element_uuid":elem[0],"transcription":elem[1],"entities_json":{"taxpayers":[]}}
    taxpayer = {"name": "","firstnames": "","activity": [],"address": [],"title": [],"familystatus": [],"birthname": ""}
    name, firstnames,activity,address,title,familystatus,birthname=[],[],[],[],[],[],[]
    for e in elem[2]:
        if e['entity_group'] == 'name+O':
            name.append(e["word"])
        if e['entity_group'] == 'firstnames+O':
            firstnames.append(e["word"])
        if e['entity_group'] == 'activity+O':
            activity.append(e["word"])
        if e['entity_group'] == 'address+O':
            address.append(e["word"])
        if e['entity_group'] == 'title+O':
            title.append(e["word"])
        if e['entity_group'] == 'familystatus+O':
            familystatus.append(e["word"])
        if e['entity_group'] == 'birthname+O':
            birthname.append(e["word"])
    taxpayer["name"] = " ".join(name)
    taxpayer["firstnames"] = " ".join(firstnames)
    taxpayer["activity"] = activity
    taxpayer["address"] = address
    taxpayer["title"] = title
    taxpayer["familystatus"] = familystatus
    taxpayer["birthname"] = " ".join(birthname)
    entry["entities_json"]["taxpayers"].append(taxpayer)
    full_json.append(entry)
```

```python
with open('taxpayers_rest.json', 'w',encoding='utf-8') as f:
    json.dump(full_json, f, ensure_ascii=False, indent=4)
```
